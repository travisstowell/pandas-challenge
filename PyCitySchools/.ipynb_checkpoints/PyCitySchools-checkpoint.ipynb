```python
# Dependencies and Setup
import pandas as pd

# File to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate Total Number of Schools
total_school_count = len(school_data_complete_df["school_name"].unique())

#Calculate Total Number of Students
total_student_count = len(school_data_complete_df["Student ID"].unique())

#Calculate Total District Budget
total_district_budget = school_data_df["budget"].sum()

#Calculate Avg Math Score
average_math_score = round(school_data_complete_df["math_score"].mean(),2)

#Calculate Avg Reading Score
average_reading_score = round(school_data_complete_df["reading_score"].mean(),2)

#Calculate # Passing Math
passing_math_count = school_data_complete_df[(school_data_complete_df["math_score"] >= 70)].count()["student_name"]

#Calculate # Passing Reading
passing_reading_count = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70)].count()["student_name"]

#Calc Percent Passing Math
percentage_passing_math = round((passing_math_count/total_student_count) * 100,2)

#Calc Percent Passing Reading
percentage_passing_reading = round((passing_reading_count/total_student_count) *100,2)

#Calc Overall Passing Percentage

school_data_complete_df.loc[school_data_complete_df['reading_score'] >= 70, 'Pass Reading'] = 1


school_data_complete_df.loc[school_data_complete_df['math_score'] >= 70, 'Pass Math'] = 1


school_data_complete_df['Pass Both'] = school_data_complete_df['Pass Reading'] + school_data_complete_df['Pass Math']


number_of_students_passing_both = (school_data_complete_df['Pass Both'].sum())/2


overall_passing = round((number_of_students_passing_both / total_student_count)*100,2)


```

```python
#creating summary data frame

district_summary_df = pd.DataFrame({"Total Schools":[total_school_count],
                                   "Total # of Students":[total_student_count],
                                   "Total Budget":[total_district_budget],
                                   "Average Math Score":[average_math_score],
                                   "Average Reading Score":[average_reading_score],
                                   "% Passing Math": [percentage_passing_math],
                                   "% Passing Reading": [percentage_passing_reading],
                                   "% Overall Passing": [overall_passing]})

district_summary_df["Total # of Students"] = district_summary_df["Total # of Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df

```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Determine the School Type & Rename Index
school_types = school_data_df.set_index(["school_name"])["type"]

# Calculate the total student count per school
per_school_counts = school_data_complete_df["school_name"].value_counts()


# Calculate the total school budget and per student spending
per_school_budget = school_data_complete_df.groupby(["school_name"]).mean()["budget"]
per_school_student = per_school_budget / per_school_counts


# Calculate the average math & reading scores
per_school_math = round(school_data_complete_df.groupby(["school_name"]).mean()["math_score"],2)
per_school_reading = round(school_data_complete_df.groupby(["school_name"]).mean()["reading_score"],2)

# Calc % Passing Math
students_passing_math = school_data_complete_df[(school_data_complete_df["math_score"]>70)]
students_passing_math_per_school = students_passing_math.groupby(["school_name"]).count()["student_name"]
percent_passing_math_schools = round((students_passing_math_per_school/per_school_counts)*100,2)


# Calc % Passing Reading
students_passing_reading = school_data_complete_df[(school_data_complete_df["reading_score"]>70)]
students_passing_reading_per_school = students_passing_reading.groupby(["school_name"]).count()["student_name"]
percent_passing_reading_schools = round((students_passing_reading_per_school/per_school_counts)*100,2)


# Calc % Passing Both
students_passing_both = school_data_complete_df[(school_data_complete_df['Pass Both'] == 2)]
students_passing_both_per_school = students_passing_both.groupby(["school_name"]).count()["student_name"]
percent_passing_both_schools = round((students_passing_both_per_school/per_school_counts)*100,2)



```

```python
#creating summary data frame

per_school_summary_df = pd.DataFrame({"School Type": school_types,
                                  "Total Students": per_school_counts,
                                  "Total School Budget": per_school_budget,
                                  "Per Student Budget": per_school_student,
                                  "Average Math Score": per_school_math,
                                  "Average Reading Score": per_school_reading,
                                  "% Passing Math": percent_passing_math_schools,
                                  "% Passing Reading": percent_passing_reading_schools,
                                  "% Overall Passing": percent_passing_both_schools})

per_school_summary_df["Total Students"] = per_school_summary_df["Total Students"].map("{:,}".format)
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.0f}".format)


per_school_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#Parse out the data by grade level using a conditional

nineth_grade = school_data_complete_df[(school_data_complete_df["grade"] == "9th")]
tenth_grade = school_data_complete_df[(school_data_complete_df["grade"] == "10th")]
eleventh_grade = school_data_complete_df[(school_data_complete_df["grade"] == "11th")]
twelfth_grade = school_data_complete_df[(school_data_complete_df["grade"] == "12th")]

#Group each grade by school
nineth_grade_scores = nineth_grade.groupby(["school_name"]).mean()["math_score"]
tenth_grade_scores = tenth_grade.groupby(["school_name"]).mean()["math_score"]
eleventh_grade_scores = eleventh_grade.groupby(["school_name"]).mean()["math_score"]
twelfth_grade_scores = twelfth_grade.groupby(["school_name"]).mean()["math_score"]

#Create combined dataframe
math_scores_by_grade_df = pd.DataFrame({"9th Avg Math Score": nineth_grade_scores, "10th Avg Math Score": tenth_grade_scores,
                                "11th Avg Math Score": eleventh_grade_scores, "12th Avg Math Score": twelfth_grade_scores})

math_scores_by_grade_df
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#Parse out the data by grade level using a conditional

nineth_grade = school_data_complete_df[(school_data_complete_df["grade"] == "9th")]
tenth_grade = school_data_complete_df[(school_data_complete_df["grade"] == "10th")]
eleventh_grade = school_data_complete_df[(school_data_complete_df["grade"] == "11th")]
twelfth_grade = school_data_complete_df[(school_data_complete_df["grade"] == "12th")]

#Group each grade by school
nineth_grade_scores = nineth_grade.groupby(["school_name"]).mean()["reading_score"]
tenth_grade_scores = tenth_grade.groupby(["school_name"]).mean()["reading_score"]
eleventh_grade_scores = eleventh_grade.groupby(["school_name"]).mean()["reading_score"]
twelfth_grade_scores = twelfth_grade.groupby(["school_name"]).mean()["reading_score"]

#Create combined dataframe
reading_scores_by_grade_df = pd.DataFrame({"9th Avg Reading Score ": nineth_grade_scores, "10th Avg Reading Score": tenth_grade_scores,
                                "11th Avg Reading Score": eleventh_grade_scores, "12th Avg Reading Score": twelfth_grade_scores})

reading_scores_by_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Bins
spending_bins = [0, 585, 630, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

#Categorize the spending based on the bins
per_school_summary_df["Spending Ranges (Per Student)"] = pd.cut(per_school_student, spending_bins, labels=group_names)

spending_math_scores = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = per_school_summary_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_rate = (spending_passing_math + spending_passing_reading) / 2

#Note to grader---for overall passing rate it looks like they are asking us to take average of average
#which doesn't seem right to me.

# Assemble into data frame
spending_summary = pd.DataFrame({"Average Math Score" : round(spending_math_scores,2),
                                 "Average Reading Score": round(spending_reading_scores,2),
                                 "% Passing Math": round(spending_passing_math,2),
                                 "% Passing Reading": round(spending_passing_reading,2),
                                 "% Overall Passing Rate": round(overall_passing_rate,2)})

spending_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Establish the bins 
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#Categorize the size based on the bins
per_school_summary_df["School Size"] = pd.cut(per_school_counts, size_bins, labels=group_names)

# Calculate the scores based on bins
size_math_scores = per_school_summary_df.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = per_school_summary_df.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = per_school_summary_df.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = per_school_summary_df.groupby(["School Size"]).mean()["% Passing Reading"]
size_overall_passing_rate = (size_passing_math + size_passing_reading) / 2

# Assemble into data frame
size_summary = pd.DataFrame({"Average Math Score" : size_math_scores,
                             "Average Reading Score": size_reading_scores,
                             "% Passing Math": size_passing_math,
                             "% Passing Reading": size_passing_reading,
                             "% Overall Passing Rate": size_overall_passing_rate})

size_summary

```
